```python
import pandas as pd
import plotly
from plotly import graph_objects as go
```

```python
data = pd.read_csv('owid-covid-data.csv')
```

```python
US_data = data[data.location=='United States']
```

```python
US_data
```

```python
traces = [
    go.Scatter(x=data[data['location']=='United States'].date, y=US_data.total_cases)]
fig = go.Figure(data=traces)
fig.show()
```
